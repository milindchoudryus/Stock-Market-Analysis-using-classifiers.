```python
#download the security data of choice (I chose google data)
import yfinance as yf
data = yf.download("GOOG", start = "2015-01-01", end = "2023-08-31")
```

```python
#check the downloaded dataset
google_data = pd.DataFrame(data)
```

```python
#create strategy1 if close price > previous close price then +1(buy), else -1(sell)
google_data['Strategy1'] = np.where(google_data['Close'].shift(-1) > google_data['Close'], 1, -1)
```

```python
#calculating the moving average of closing prices 1. 50 days average (MA50), 2. 200 days average (MA200)
google_data['MA50'] = google_data['Close'].rolling(window = 50).mean()
google_data['MA200'] = google_data['Close'].rolling(window = 200).mean()
```

```python
#create strategy 2 if MA50 > MA200 then buy(1), else sell (-1)
google_data['Strategy2'] = np.where(google_data['MA50'] > google_data['MA200'], 1, -1)
```

```python
#drop and clean the ambiguish data from the dataset (clean the data)
google_data = google_data.dropna()
google_data
```

```python
#import few more models
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler as scaler
```

```python
#label the feature variale 'X', and target variable 'y'

X_strategy1 = google_data[['Close']]
y_strategy1 = google_data['Strategy1']


X_strategy2 = google_data[['MA50','MA200']]
y_strategy2 = google_data['Strategy2']
```

```python
#split the data into training and testing dataset (80-20 split)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_strategy1, y_strategy2, test_size = 0.2, random_state = 42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_strategy2, y_strategy2, test_size = 0.2, random_state = 42)
```

```python
# Choose a Classifier (KNN) and fit it on the training dataset (default parameters)
clf1 = KNeighborsClassifier(n_neighbors = 10)
clf1.fit(X_train1, y_train1)

clf2 = KNeighborsClassifier(n_neighbors = 10)
clf2.fit(X_train2, y_train2)

# Evaluate and compare the classifiers on the test dataset
y_pred1 = clf1.predict(X_test1)
accuracy1 = accuracy_score(y_test1, y_pred1)

y_pred2 = clf2.predict(X_test2)
accuracy2 = accuracy_score(y_test2, y_pred2)
```

```python
#printing the accuracy score (As this is a classification measure which is used to evalute the correctness of the preditions)
print(accuracy1)
print(accuracy2)
```

**Note:** As per the research accuracy_score is good if it is in between 70%-90% i.e., 0.7 - 0.9. And it's best if the score is > 0.9. **bold text**

```python
#choose a classifier (Random Forest) and fit it on the training dataset (default parameters)
rclf1 = RandomForestClassifier(random_state = 42)
rclf1.fit(X_train1, y_train1)

rclf2 = RandomForestClassifier(random_state = 42)
rclf2.fit(X_train2, y_train2)

# Evaluate and compare the classifiers on the test dataset
y_pred1 = rclf1.predict(X_test1)
r_accuracy1 = accuracy_score(y_test1, y_pred1)

y_pred2 = rclf2.predict(X_test2)
r_accuracy2 = accuracy_score(y_test2, y_pred2)

print(r_accuracy1)
print(r_accuracy2)
```

```python
# Choose a Classifier (Gradient Boosting) and fit it on the training dataset (default parameters)
from sklearn.ensemble import GradientBoostingClassifier

gclf1 = GradientBoostingClassifier(random_state = 42)
gclf1.fit(X_train1, y_train1)

gclf2 = GradientBoostingClassifier(random_state = 42)
gclf2.fit(X_train2, y_train2)

# Evaluate and compare the classifiers on the test dataset
y_pred1 = gclf1.predict(X_test1)
g_accuracy1 = accuracy_score(y_test1, y_pred1)

y_pred2 = gclf2.predict(X_test2)
g_accuracy2 = accuracy_score(y_test2, y_pred2)

print(g_accuracy1)
print(g_accuracy2)
```
